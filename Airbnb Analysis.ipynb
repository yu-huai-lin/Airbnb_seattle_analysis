### 1. Convert true,false to 1 and 0

```python
def convert_tf(df, columns, true_value = 't'):
    for c in columns:
        df[c] = [1 if i==true_value else 0 for i in df[c]]
```

```python
convert_tf(df, ['host_is_superhost','is_location_exact', 'has_availability','requires_license', 'instant_bookable',  'require_guest_profile_picture','require_guest_phone_verification'],
          't')
```
